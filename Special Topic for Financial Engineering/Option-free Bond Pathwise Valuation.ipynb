```python
import numpy as np
import pandas as pd

interest = {
    "Year1": [0.025000, 0.025000, 0.025000, 0.025000, 0.025000, 0.025000, 0.025000, 0.025000],
    "Year2": [0.045160, 0.045160, 0.045160, 0.045160, 0.033456, 0.033456, 0.033456, 0.033456],
    "Year3": [0.067332, 0.067332, 0.049880, 0.049880, 0.049880, 0.049880, 0.036952, 0.036952],
    "Year4": [0.091081, 0.067474, 0.067474, 0.049986, 0.067474, 0.049986, 0.049986, 0.037031]
}

n = 3

index = ["Path" + str(i+1) for i in range(2**n)]
column = ["Year" + str(i+1) for i in range(n+1)]

pathwise = pd.DataFrame(interest, index=index)
pathwise
```

```python
face_value = 100
coupon_rate = 0.0275
c = face_value * coupon_rate

pathwise["Value"] = (c/(1+pathwise["Year1"])) + \
                    (c/( (1+pathwise["Year1"]) * (1+pathwise["Year2"]) )) + \
                    (c/( (1+pathwise["Year1"]) * (1+pathwise["Year2"]) * (1+pathwise["Year3"]) )) + \
                    ((face_value + c)/( (1+pathwise["Year1"]) * (1+pathwise["Year2"]) * (1+pathwise["Year3"]) * (1+pathwise["Year4"]) ))
pathwise
```

```python
print("The average value of bond is $" + str(round(pathwise["Value"].mean(),2)))
```
