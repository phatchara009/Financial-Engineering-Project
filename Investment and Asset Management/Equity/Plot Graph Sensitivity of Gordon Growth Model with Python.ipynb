```python
# Import package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create Variable
d0 = 1.8
r_1 = 0.15
g_1 = np.arange(0.015, 0.125, 0.01)

# Create DataFrame
dict_x = {"d0":d0, "g":g_1, "r":r_1}
data_x = pd.DataFrame.from_dict(dict_x)

# Create Value of common stock column
data_x["v"] = (data_x["d0"] * (1 + data_x["g"])) / (data_x["r"] - data_x["g"])
print(data_x)

# Plot graph
plt.plot(data_x["g"], data_x["v"])
plt.title("Sensitivity of V and g")
plt.xlabel("Growth (g)")
plt.ylabel("Value of Stock (V)")
plt.show()
```

```python
# Create Variable
d0 = 1.8
r_2 = np.arange(0.05, 0.16, 0.01)
g_2 = 0.035

# Create DataFrame
dict_y = {"d0":d0, "g":g_2, "r":r_2}
data_y = pd.DataFrame.from_dict(dict_y)

# Create Value of common stock column
data_y["v"] = (data_y["d0"] * (1 + data_y["g"])) / (data_y["r"] - data_y["g"])
print(data_y)

# Plot graph
plt.plot(data_y["r"], data_y["v"])
plt.title("Sensitivity of V and r")
plt.xlabel("Required Rate of Return (r)")
plt.ylabel("Value of Stock (V)")
plt.show()
```
