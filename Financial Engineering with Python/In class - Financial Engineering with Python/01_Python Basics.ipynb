##3.1 If Statement
An if statement is a conditional statement that runs or skips code based on whether a condition is true or false.

```python
saving_balance = 15
if saving_balance < 15:
    saving_balance += 10
else:
    saving_balance -= 10
print(saving_balance)
```

```python
season = "summer" #there is no difference in single or double quoted string
if season == 'spring':
    print('plant the garden!')
elif season == 'summer':
    print('water the garden!')
elif season == 'fall':
    print('harvest the garden!')
elif season == 'winter':
    print('stay indoors!')
else:
    print('unrecognized season')
```

Be careful with the order of operations! See Section 6.17. Operator precedence at https://docs.python.org/3/reference/expressions.html#summary for reference.

**Good and Bad Examples of boolean expressions**

* Don't use True or False as conditions

```python
# Bad example
if True:
    print("This indented code will always get run.")
```

* Be careful writing expressions that use logical operators.
* In Python, any non-zero and non-empty value is considered "truthy", which means it evaluates to True in a boolean context


```python
weather = "snow"

# Bad example
if weather == "snow" and "cloudy": # "cloudy", is not a boolean expression - it's a string (and is always True)
    print("Wear boots!")
else:

    print("Wear sandals!")
```

* Don't compare a boolean variable with == True or == False

```python
# Bad example
is_cold = False
if is_cold == True:
    print("The weather is cold!")
else:
    print("The weather is warm!")
```

### Useful functions

**The zip()** function is used to combine two or more iterables (e.g., lists, tuples) into an iterator of tuples. Each tuple contains the corresponding elements from the input iterables.

```python
days = [1, 2, 3, 4, 5]
daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02]

zipped_data = zip(days, daily_returns)
for day, return_value in zipped_data:
    print(day, return_value)
```

The number of iterations will be equal to the length of the shortest input iterable.

```python
days = [1, 2, 3, 4, 5, 6 ] # length = 6
daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02] # length = 5

zipped_data = zip(days, daily_returns)
for day, return_value in zipped_data:
    print(day, return_value)
```

* If you want to iterate through both lists, even if one is shorter than the other, you can use the **itertools**.

* zip_longest() function from the itertools module. This function fills in missing values with a specified fill value (defaulting to None).

```python
from itertools import zip_longest

days = [1, 2, 3, 4, 5, 6] # length = 6
daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02] # length = 5

for day, return_value in zip_longest(days, daily_returns):
    print(day, return_value)
```

##4.2 While loop
* For loops are an example of "definite iteration" meaning that the loop's body is run a predefined number of times.
* Whereas while loops repeat an unknown number of times and ends when a condition is not true.

```python
card_deck = [4, 11, 8, 5, 13, 2, 10]
hand = []

# adds the last element of the card_deck list to the hand list
# until the values in hand add up to 17 or more
while sum(hand) < 17:
    pop = card_deck.pop() # take the last element out of the list
    print("pop={}".format(pop))
    hand.append(pop)
    print("sum={}".format(sum(hand)))
    print()

print ("The result from while loop")
print("card_deck = {}".format(card_deck))
print("hand = {}".format(hand))
```

**Exercise 2** -  Use a while loop to calculate the cumulative return from a series of daily returns.

daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02]

Expected result:

Cumulative Return =  0.03

```python
daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02]
cumulative_return = 1.0  # Initial value
i = 0

# Use a while loop to calculate a cumulative return
while i < len(daily_returns):
    cumulative_return *= (1 + daily_returns[i])
    i += 1

# Print the result
print(round(cumulative_return-1, 2))
```

```python
cylinder_volume(10, 7)  # pass by position
cylinder_volume(radius=7, height=10)  # pass by name (no need to follow the order of arguments)
```

**Optional Exercise** Calculate Market Capitalization Density

In this exercise, you will create a Python function named market_capitalization_density that calculates the market capitalization density of a financial instrument (e.g., stocks, bonds). Market capitalization density is the market capitalization divided by the total number of financial instruments available.

The formula to calculate the market capitalization density is:



$market\_capitalization\_density = market\_capitalization / total\_instruments$



If total_instruments not provided, the default value should be 1000.



If Total market capitalization is 10,000.00 USD and no total_instruments provided,  the market capitalization density will be 1000.00 USD/instrument


```python
## Define your function here
def market_capitalization_density(market_capitalization,
                                  total_instruments = 1000):
    market_capitalization_density = market_capitalization / total_instruments
    return market_capitalization_density

# Test and print your function
total_market_cap = 1000000  # Total market capitalization in USD
market_capitalization_density(market_capitalization = total_market_cap)
```

##5.2 Variable Scope

* If a variable is created inside a function, it can only be used within that function.

```python
# This will result in an error
def some_function():
  word = "hello"
  print(word)


print(word)
```

```python
# This works fine
def some_function():
    word = "hello"
    print(word)

some_function()
```
