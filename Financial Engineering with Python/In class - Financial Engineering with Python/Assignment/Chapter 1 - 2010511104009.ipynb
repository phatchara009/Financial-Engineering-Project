**Exercise 2** -  Use a while loop to calculate the cumulative return from a series of daily returns.

daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02]

Expected result:

Cumulative Return =  0.03

```python
daily_returns = [0.02, -0.01, 0.03, 0.01, -0.02]
cumulative_return = 1.0  # Initial value
i = 0

# Use a while loop to calculate a cumulative return
while i < len(daily_returns):
    cumulative_return *= (1 + daily_returns[i])
    i += 1

# Print the result
print(round(cumulative_return - 1, 2))
```

Create a function to calculate a future value. Your function should have 3 arguments: present value, interest rate, and year.

If pv=100 (present value), year=3, rate=0.015

FV should be 104.56783749999997


```python
## Write your function here
def calculate_future_values(pv, int_rate, year):
    fv = pv * ((1 + int_rate) ** year)
    return fv

## call your function and print the result
print(calculate_future_values(100, 0.015, 3))
```

**Optional Exercise** Calculate Market Capitalization Density

In this exercise, you will create a Python function named market_capitalization_density that calculates the market capitalization density of a financial instrument (e.g., stocks, bonds). Market capitalization density is the market capitalization divided by the total number of financial instruments available.

The formula to calculate the market capitalization density is:



$market\_capitalization\_density = market\_capitalization / total\_instruments$



If total_instruments not provided, the default value should be 1000.



If Total market capitalization is 10,000.00 USD and no total_instruments provided,  the market capitalization density will be 1000.00 USD/instrument


```python
## Define your function here
def market_capitalization_density(market_capitalization,
                                  total_instruments = 1000):
    market_capitalization_density = market_capitalization / total_instruments
    return market_capitalization_density

# Test and print your function
total_market_cap = 1000000  # Total market capitalization in USD
print(market_capitalization_density(market_capitalization = total_market_cap))
```
