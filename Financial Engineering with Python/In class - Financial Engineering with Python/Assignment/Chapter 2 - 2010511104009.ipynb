**Exercise 4**

Build a linear regression model for the following data.

x=[1,2,3,4,5,6,7]

y=[1,2,3,4,2,3,4]

Then create a plot showing a line of (x,y) and (x, predicted_y).

Expected Result

image.png


```python
import matplotlib.pyplot as plt
from scipy import stats

y =[10,20,30,20,20,30,40]
x =[1,2,3,4,5,6,7]

# fit a model and get a beta and alpha values
(beta, alpha, r_value, p_value, std_err) = stats.linregress(x,y) #no need to manually add constant as in statsmodels

# predict the y using the beta and alpha
predicted_y = alpha + (beta * np.array(x))

# make a plot containing 2 lines of (x,y) and (x, predicted_y)
plt.plot(x, y, label = "actual y")
plt.plot(x, predicted_y, label = "predicted y")
plt.legend()
plt.show()
```
