```
# Import library
import pandas as pd 
import numpy as np
import os
!pip install bar_chart_race
import bar_chart_race as bcr

# Download Data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/countries-aggregated_csv.csv')

# Check head of data 5 row
df.head()

# Check information of data
df.info()

# Check descriptive statistics
df.describe()

# Focus ASEAN nation
sea_list = ['Brunei','Cambodia','Indonesia','Laos','Malaysia','Burma','Philippines','Singapore','Thailand','Vietnam']
print(sea_list)
sea_df = df[df['Country'].isin(sea_list)]
sea_df = sea_df[['Date','Country','Confirmed']]
print(sea_df)

# Using pivot func to format data to show for one value of nation
sea_df2 = sea_df.pivot(index='Date', columns='Country', values='Confirmed')
print(sea_df2)

# Plot animated bar chart race!!!
bcr.bar_chart_race(df=sea_df2, filename=None,period_length=200, figsize = (3.5,3),title='COVID-19 Confirmed cases in ASEAN')
```
